### calculating feature importances using Random Forest Classifier algorithm

```python
importances = np.zeros(X.shape[1])
count = 0

for i in tqdm(range(100)):
    seed = random.randint(1, 10000)
    random.seed(seed)
    test_size = round(random.randint(10, 40)/100, 2)
    x_train, x_test, y_train, y_test = train_test_split(X, y, random_state=seed, test_size=test_size)
    
    rf_classifier = RandomForestClassifier(n_estimators=random.randint(100, 400), random_state=seed)
    rf_classifier.fit(x_train, y_train)
    y_pred = rf_classifier.predict(x_test)
    accuracy = accuracy_score(y_test, y_pred)
    if accuracy>0.9:
        importances += rf_classifier.feature_importances_
        count+=1
        
importances = importances/count
print("Number of Instances with model accuracy >90: ", count)

feature_indices = sorted(range(len(importances)), key=lambda k: importances[k], reverse=True)

print("Feature Importances:")
for idx in feature_indices[:20]:
    print(f"Feature {idx + 1}: {importances[idx]}")
```
